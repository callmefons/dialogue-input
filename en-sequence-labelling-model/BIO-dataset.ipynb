```python
import pandas as pd
import re
import spacy
import numpy as np
from spacy import displacy
from spacy.tokenizer import Tokenizer
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english')) 
```

```python
nlp = spacy.load("en_core_web_sm")
med7 = spacy.load("en_core_med7_lg")
```

```python
data = pd.read_csv("data/nursing-noteevents.csv")
data.head()
```

```python
new_sentences = []
count_ent = []
for index, row in data.iterrows():
    text = row['text']
    actions = re.split(r'\s(?=\w+:)', text)
    for action in actions:
        if(action.find("Action:") != -1 and action.find("Response:")):
            doc = med7(action)
            if len(doc.ents) > 0:
                new_sentences.append(action.split("Action:\n")[1].strip().split("Response:\n")[0])
                count_ent.append(len(doc.ents))
                print(action)
                print(doc.ents)
```

```python
new_sentences = pd.DataFrame(new_sentences, columns = ['Sentence']) 
new_sentences.to_csv("data/drug.csv")
```

```python
pos_tag = []
def pos_tagging(text):
    t = ""
    pos = []
    word_tok = []
    doc = nlp(text)
    for token in doc:
#         print(token.text, token.pos_)
        t += token.text + " "
        word_tok.append(token.text)
        pos.append(token.pos_)
        pos_tag.append(token.pos_)
    return t, word_tok, pos
```

```python
def bi_tagging(text):
    doc = med7(text)

    tup_ents = []
    ents = []
    sent = ""

    for ent in doc.ents:
        for i, token in enumerate(word_tokenize(str(ent))):
            if i == 0:
                tup_ents.append((token, "B-"+str(ent.label_)))
            else:
                tup_ents.append((token, "I-"+str(ent.label_)))
            ents.append(token)
            sent += token + " "

    return tup_ents, ents, sent
```

```python
def o_tagging(text, ents):
    tup_tokens = []
    tokens = []
    sent = ""
    for token in word_tokenize(text):
        if token not in ents:
            tokens.append(token)
            sent += token + " "
            tup_tokens.append((token, "O"))

    return tup_tokens, tokens, sent
```

```python
sentences = []
for idx, text in enumerate(new_sentences):
    t, word_tok, pos = pos_tagging(text)
    tup_ents, ents, ent = bi_tagging(t)
    tup_tokens, tokens, token = o_tagging(text, ents)
    tups = tup_tokens + tup_ents
    
    for idy, word in enumerate(word_tok):
        for tup in tups:
            if word == tup[0]:
                sent = "Sentence: "+ str(idx)
                tag = tup[1]
                if pos[idy]:
                    sentences.append([sent, word, pos[idy], tag])
                else:
                    sentences.append([sent, word, "", tag])
    
# sentences
```

```python
df = pd.DataFrame(sentences, columns = ['Sentence #', 'Word', "POS", "Tag"]) 
df.head() 
```
